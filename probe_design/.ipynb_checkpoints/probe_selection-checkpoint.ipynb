```python
# libraries

import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load tables

gene_exp = pd.read_csv('SW1_SW11_Counts.csv',dtype=None)

#gene_exp.shape
gene_exp
#gene_exp.dtype
#gene_exp[0][0]
```

```python
# calculate day 0 and day 28 averages

d0_vals = gene_exp[['SW.1.d0.1','SW.1.d0.2','SW.1.d0.3','SW.11.d0.1','SW.11.d0.2','SW.11.d0.3']]
d28_vals = gene_exp[['SW.1.d28.1','SW.1.d28.2','SW.1.d28.3','SW.11.d28.1','SW.11.d28.2','SW.11.d28.3']]


d0_avg = np.mean(d0_vals,axis=1)
d28_avg = np.mean(d28_vals,axis=1)

plt.figure
plt.scatter(d0_avg,d28_avg)
plt.xscale("log")
plt.yscale("log")

axes = plt.gca()
axes.set_xlim([1,max(d0_avg)])
axes.set_ylim([1,max(d28_avg)])
```

```python
# threshold genes

thresh = 1000

all_genes = gene_exp['external_gene_name']
sel_index = ((d0_avg > thresh) & (d28_avg > thresh))

sel_genes = all_genes[sel_index].reset_index(drop=True)
sel_genes
```

```python
# find SNPs in highly expressed exons

snps_in_exons = pd.read_csv('snps_in_exons.txt',sep="\t",header=None)
sel_snps_index = np.isin(snps_in_exons[[2]],sel_genes)
sel_snps_in_exons = snps_in_exons[sel_snps_index]

#np.flatnonzero(sel_snps_index)
```
